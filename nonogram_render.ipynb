```python
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
```

```python
def get_line(board):
    ret_line = []

    for row in board:
        line = []
        x_cnt = 0
        for index, elem in enumerate(row):
            if elem == 1:
                x_cnt += 1
                if index == row.size - 1:
                    line.append(x_cnt)
            else:
                if x_cnt != 0:
                    line.append(x_cnt)
                x_cnt = 0
        ret_line.append(line)


    return ret_line

def generator(size):
    board = np.random.randint(0, 2, size=(size, size))
    xl = get_line(board)
    board = board.T
    yl = get_line(board)
    board = board.T

    return xl, yl, board, size
```

```python
def render_board(x_label, y_label, board, size):

    # create discrete colormap
    cmap = colors.ListedColormap(['white', 'black'])

    fig, ax = plt.subplots(dpi=600)
    ax.imshow(board, cmap=cmap)
    plt.title("NONOGRAM")
    xl = []
    yl = []

    for s in range(size):
        ax.axhline(s - 0.5, color='black')
        ax.axvline(s - 0.5, color='black')

    for elem in x_label:
        x = list(map(str, elem))
        xl += ["\n".join(x)]

    for elem in y_label:
        y = list(map(str, elem))
        yl += [" ".join(y)]

    plt.xticks(np.arange(0, size), xl)
    plt.yticks(np.arange(0, size), yl)

    ax.tick_params(labeltop=True, labelbottom=False, direction="out")
    #plt.savefig('monogram.png', dpi=600)
    plt.show()
```

```python
xl, yl, board, size = generator(15)
render_board(yl, xl, board, size)
```
